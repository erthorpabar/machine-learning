```python
''' 
信息熵


0 基本概念
m = 可能的选项(信息量)
p = 选项的概率(概率)

比如 抛硬币
p(a1) = 1/2
p(a2) = 1/2

m=2代表有两个选项(概率的最大公约数)
mp 
= m*p(a1) + m*p(a2) 
= 2*0.5 + 2*0.5
= 1


1 抛一次硬币 n = 1
p(a1) = 1/2
p(a2) = 1/2
最高效率 每次排除一半信息量 需要1次确认结果
整体信息熵n = log2(m=2) = 1 bit
m = 1/p
所以 n = log2(1/p)

对于单个事件的信息熵
n(a1) = p * log2(1/p) = 1/2 * log2(2) = 0.5 bit
n(a2) = p * log2(1/p) = 1/2 * log2(2) = 0.5 bit
整体信息熵 n = n(a1) + n(a2) = 1 bit


2 抛两次硬币 n = 2
p(a1) = 1/4
p(a2) = 1/4
p(a3) = 1/4
p(a4) = 1/4
最高效率 每次排除一半信息量 需要2次确认结果
整体信息熵 n = log2(m=4) = 2 bit
m = 1/p
所以 n = log2(1/p)

对于单个事件的信息熵
n(a1) = p * log2(1/p) = 1/4 * log2(4) = 0.5 bit
n(a2) = p * log2(1/p) = 1/4 * log2(4) = 0.5 bit
n(a3) = p * log2(1/p) = 1/4 * log2(4) = 0.5 bit
n(a4) = p * log2(1/p) = 1/4 * log2(4) = 0.5 bit
整体信息熵 n = n(a1) + n(a2) + n(a3) + n(a4) = 2 bit


3 总结
0基础概念
m = 信息量(可选项数量)(概率的最大公约数)
p = 选项的概率
n = 信息熵
mp = 1

1 信息熵
单个事件的信息熵 n(i) = log2(m) = log2(1/p) 
整个事件的信息熵 n = sum(p(i) * n(i)) = sum(p * log2(1/p)) = sum(p * log2(m))

'''
```

```python
''' 
非等概率的信息熵

1 
p(a1) = 4/8
p(a2) = 2/8
p(a3) = 1/8
p(a4) = 1/8

m = 8
单个事件的信息熵
n(i) = p * log2(1/p)
n(a1) = p * log2(1/p) = 4/8 * log2(8/4) = 1/2 bit
n(a2) = p * log2(1/p) = 2/8 * log2(8/2) = 1/4 bit
n(a3) = p * log2(1/p) = 1/8 * log2(8/1) = 3/8 bit
n(a4) = p * log2(1/p) = 1/8 * log2(8/1) = 3/8 bit
整体信息熵 n = n(a1) + n(a2) + n(a3) + n(a4) = 1.75 bit


此时n = 1.75 bit
当所有概率相等的时候 代表无区分度 则 熵n最大
即 等概率的时候 熵最大
n(max) = log2(m) = log2(8) = 3 bit


2 另一种计算信息熵的方式
p(a1) = 1/2
p(a2) = 1/4
p(a3) = 1/8
p(a4) = 1/8

想象一个决策树
   / \
a1   / \
   a2  / \
      a3  a4

n = 3 次决策
当第n= log2(m=2) = 1 次决策时 分类出a1
当第n= log2(m=4) = 2 次决策时 分类出a2
当第n= log2(m=8) = 3 次决策时 分类出a3,a4

n = 决策总次数
m = 排除的选项数量
n(i)/m(i) = 每次判断 排除了多少选项
信息熵 n = 平均识别一个事件所需的最小判断次数
信息熵 n ≈ 决策树中的平均路径长度

n1 = 1
n2 = 2
n3 = 3
n4 = 3

m1(max) = 2
m2(max) = 2^2 = 4
m3(max) = 2^3 = 8
m4(max) = 2^3 = 8

n = sum(n(i)/m(max)) 
= n1/m1 + n2/m2 + n3/m3 + n4/m4
= 1/2 + 2/4 + 3/8 + 3/8
= 1.75

p(平均) = 1/m(max)
p1 = 1/2
p2 = 1/4
p3 = 1/8
p4 = 1/8

n = sum(n(i)*p(平均))
= n1*p1 + n2*p2 + n3*p3 + n4*p4
= 1 * 1/2 + 2 * 1/4 + 3 * 1/8 + 3 * 1/8
= 1.75


3 总结
0基础概念
m = 信息量(可选项数量)(概率的最大公约数)
p = 选项的概率
n = 信息熵 = 得到结果的平均判断次数
mp = 1

1等概率
整个事件的最大信息熵(等概率) n(max) = log2(m) = log2(1/p) 

2非等概率(通用)
2-1
单个事件的信息熵 n(i) = log2(m) = log2(1/p) 
整个事件的信息熵 n = sum(p(i) * n(i)) = sum(p * log2(1/p)) = sum(p * log2(m))

2-2 
必须先有决策树n(i) 才能计算信息熵
n = sum(n(i)/m(max)) = sum(n(i)*p(平均))

'''
```

```python
''' 
1交叉熵

正确分类的熵
 / \
A  / \
   B  C

n(A) = 1
n(B) = 2
n(C) = 2

p(A) = 0.5
p(B) = 0.25
p(C) = 0.25

信息熵 n
= n(A) * p(A) + n(B) * p(B) + n(C) * p(C)
= 1 * 0.5 + 2 * 0.25 + 2 * 0.25
= 1.5


错误分类的熵(弄错了决策路径,但是概率分布是不变的,导致平均决策次数增加)
 / \
C  / \
   B  A

n(A错) = 2
n(B错) = 2
n(C错) = 1

p(A错) = 0.25
p(B错) = 0.25
p(C错) = 0.5

交叉熵 h
= n(A错) * p(A) + n(B错) * p(B) + n(C错) * p(C)
= 2*0.5 + 2*0.25 + 1*0.25
= 1.75

n = log2(1/p)
使用p计算交叉熵h
= p(A) * log2(1/p(A错)) + p(B) * log2(1/p(B错)) + p(C) * log2(1/p(C错))
= 0.5 * log2(1/0.25) + 0.25 * log2(1/0.25) + 0.25 * log2(1/0.5)
= 0.5 * 2 + 0.25 * 2 + 0.25 * 1
= 1.75

2 相对熵
相对熵 = 交叉熵 - 信息熵


3 总结
0基础概念
m = 信息量(可选项数量)(概率的最大公约数)
p = 选项的概率
n = 信息熵 = 得到结果的平均判断次数(好的决策树 应优先考虑最具区分度的判断条件)
mp = 1

1 最大熵
= 1无任何决策依据 2 各种可能无任何区分度 3 平均判断次数最多 4 信息熵最大
整个事件的最大信息熵(等概率) 
n(max) = log2(m) = log2(1/p) 

2 最小熵 = 非等概率熵
= 1 最优决策依据 2 各选项有最佳区分度 3 平均决策次数最小 4 信息熵最小
2-1
单个事件的信息熵 n(i) = log2(m) = log2(1/p) 
整个事件的信息熵 n = sum(p(i) * n(i)) = sum(p * log2(1/p)) = sum(p * log2(m))

2-2 
必须先有决策树n(i) 才能计算信息熵
n = sum(n(i)/m(max)) = sum(n(i)*p(平均))


3 实际熵 = 交叉熵 = 当前决策树的熵
= 1 非最优决策依据 2 各选项非最佳区分度 3 平均决策次数介于最大最小之间 4 熵介于最大最小之间

4 相对熵 = 交叉熵 - 信息熵



4 举例子
最大熵
  / \
/ \  / \
A  A B  C
n(max) 
= p(A)*n(A) +p(A)*n(A) + p(B)*n(B) + p(C)*n(C)
= 0.25*2 + 0.25*2 + 0.25*2 + 0.25*2
= 2

最小熵
 / \
A  / \
   B  C
n(min) 
= p(A)*n(A) + p(B)*n(B) + p(C)*n(C)
= 0.5*1 + 0.25*2 + 0.25*2
= 1.5

交叉熵
 / \
C  / \
   B  A
n(cross)
= p(A)*n(A错) + p(B)*n(B错) + p(C)*n(C错)
= 0.25*2 + 0.25*2 + 0.5*1
= 1.75

相对熵
= n(cross) - n(min)
= 1.75 - 1.5
= 0.25

'''
```
