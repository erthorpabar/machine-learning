```python
''' svd 降维
————————数学原理————————
0 核心思想
找到权重最大的特征 并且筛选
(m,n) = (m,r) * (r,r) * (r,n)
从r个特征中选择k个特征
(m,n) ≈ (m,k) * (k,r) * (r,n)


1 数据
A (m,n) m 样本数量 n 特征

2 A.T * A 的特征值 特征向量
A.T * A = V * λ * V.T
V = 特征向量 = 右奇异向量
λ = 特征值 = a^2(奇异值的平方)
 
3 A * A.T 的特征值 特征向量
A * A.T = U * λ * U.T
U = 特征向量 = 左奇异向量


4 奇异值a
a = √λ
r = 按照a列表构建的(m,n)对角线矩阵

5 SVD分解
A = U * r * V.T

6 选择k个特征

————————案例————————

1 数据
A (3,2) 样本数量 m=3 n=2
A = [[1, 2],
     [3, 4],
     [5, 6]]

2 A.T * A 的特征值 特征向量
A.T *A = [[35, 44],
        [44, 56]]

特征值 λ = [90.7, 0.3]
特征向量 v = [[-0.64, -0.77],
              [-0.77,  0.64]]



3 A * A.T 的特征值 特征向量
A * A.T  = [[5,  11, 17],
            [11, 25, 39], 
            [17, 39, 61]]

特征向量 U = [[-0.23, 0.89, 0.40],
              [-0.52, 0.24, -0.82],
              [-0.82, -0.41, 0.41]]


4 奇异值
a = √λ = [√90.7, √0.3] = [9.5, 0.5]
r = [[9.5, 0  ],
     [0,   0.5],
     [0,   0  ]]

     
5 SVD分解
A = U * r * V.T
(3,2) = (3,3) * (3,2) * (2,2)
     
     
6 选择k个特征
(m,n) ≈ (m,k) * (k,r) * (r,n)

k = 1
U(k=1)= [[-0.23],
    [-0.52], 
    [-0.82]]
r(k=1) = = [[9.5]]
v(k=1).T = [[-0.64, -0.77]]

A' = U * r * v.T
(3,2) ≈ (3,1) * (1,1) * (1,2)
'''
```
